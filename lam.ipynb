# Xây dựng mô hình dự báo giá đóng cửa trong phiên giao dịch kế tiếp

```python
from sklearn.linear_model import LinearRegression
# Import thu vien LinearRegression
```

# Lấy giá đóng cửa và khối lượng giao dịch của ngày cuối cùng

```python
last_close = df_vcb.iloc[-1]["Close"]
last_close
```

```python
last_volume = df_vcb.iloc[-1]["Volume"]
last_volume
```

```python
# Chuẩn bị dữ liệu cho model
X = df_vcb[["Close", "Volume"]].iloc[:-1] # Dữ liệu feature (giá đóng cửa, khối lượng giao dịch)
y = df_vcb["Close"].shift(-1).iloc[:-1] # Dữ liệu target (giá đóng cửa ngày kế tiếp)

# Tạo model Linear Regression và fit dữ liệu
model = LinearRegression()
model.fit(X, y)
```

```python
# Tạo input cho việc dự báo giá đóng cửa của phiên giao dịch kế tiếp
input_data = [[last_close, last_volume]]

# Dự báo giá đóng cửa của phiên giao dịch kế tiếp
predicted_close = model.predict(input_data)
print("Giá đóng cửa dự báo của phiên giao dịch kế tiếp:", predicted_close[0])
```

# Vẽ biểu đồ so sánh giá dự báo và giá thực tế

```python
# Tạo biểu đồ
plt.figure(figsize=(12, 6))
plt.plot(df_vcb["Close"], label="Giá thực tế")
plt.plot(df_vcb.index[-1] + timedelta(days=1), predicted_close[0], 'r.', markersize=15, label="Giá dự báo")
plt.ylabel("Giá cổ phiếu")
plt.xlabel("Ngày")
plt.legend()
plt.show()
```

```python
# Tính toán các loại sai số
mae = mean_absolute_error(df_vcb["Close"].shift(-1).iloc[:-1], model.predict(X))
mse = mean_squared_error(df_vcb["Close"].shift(-1).iloc[:-1], model.predict(X))
rmse = mean_squared_error(df_vcb["Close"].shift(-1).iloc[:-1], model.predict(X), squared=False)
mape = np.mean(np.abs((df_vcb["Close"].shift(-1).iloc[:-1] - model.predict(X)) / df_vcb["Close"].shift(-1).iloc[:-1])) * 100
R2 = r2_score(df_vcb["Close"].shift(-1).iloc[:-1], model.predict(X))

# Hiển thị kết quả
print("Mean absolute error (MAE):", mae)
print("Mean squared error (MSE):", mse)
print("Root mean square error (RMSE):", rmse)
print("Mean absolute percent error (MAPE):", mape)
print("R-squared (R2):", R2)
```
